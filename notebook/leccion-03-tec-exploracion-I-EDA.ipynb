# Conceptos importantes 

En este apartado vamos a definir algunos de los conceptos que más repetiremos a lo largo de todo el módulo y que es importante que tengamos claros:

- **Variables**:  una variable se refiere a una característica o atributo específico que se está midiendo, registrando o estudiando en un conjunto de datos. Puede ser cualquier cantidad, calidad o propiedad que pueda variar entre diferentes elementos o unidades de observación en un estudio. Las variables pueden clasificarse en diferentes tipos, y el tipo de variable determina las operaciones que se pueden realizar con ella y las formas en que se puede analizar. En nuestro caso, cada variable será cada columna de nuestro *DataFrame*. 

- **Tipos de variables**: Es importante comprender los tipos de variables presentes en un conjunto de datos, ya que pueden afectar el tipo de análisis y las técnicas que se pueden aplicar a los datos.

    - **Numéricas**: Estas variables representan valores numéricos y se utilizan para medir magnitudes. En el conjunto de datos, la variable "age" (edad) es un ejemplo de una variable numérica.

    - **Categóricas**: Estas variables representan categorías o etiquetas discretas. En el conjunto de datos, las variables "job" (tipo de trabajo), "marital" (estado civil), "education" (educación)-, "default" (crédito en incumplimiento), "housing" (préstamo hipotecario), "loan" (préstamo personal), "contact" (tipo de comunicación de contacto), "month" (mes), "day_of_week" (día de la semana) y "poutcome" (resultado de la campaña anterior) son ejemplos de variables categóricas.Dentro de las variables categóricas, podemos tener varios tipos dependiendo de la naturaleza de los datos. A continuación, se presentan algunos tipos comunes de variables categóricas:

        - Nominales: Son variables categóricas sin un orden inherente entre las categorías. Las categorías representan diferentes grupos o clases, pero no hay una relación de jerarquía entre ellas. Ejemplos de variables nominales en el conjunto de datos proporcionado podrían ser "job" (tipo de trabajo) y "marital" (estado civil).

        - Ordinales: Son variables categóricas con un orden o jerarquía entre las categorías. Las categorías representan diferentes niveles o rangos, y el orden entre ellas es significativo. Ejemplos de variables ordinales en el conjunto de datos podrían ser "education" (educación), donde las categorías pueden indicar niveles educativos ascendentes (por ejemplo, 'basic.4y', 'basic.6y', 'basic.9y', 'high.school', 'professional.course', 'university.degree'), o "poutcome" (resultado de la campaña anterior), donde las categorías pueden ser 'failure', 'nonexistent' y 'success'.

        - Binarias: Son un caso especial de variables categóricas que toman solo dos categorías posibles. Las categorías se pueden representar como '0' y '1', 'no' y 'yes', o cualquier otra combinación de dos valores opuestos. Un ejemplo de variable binaria en el conjunto de datos es "default" (crédito en incumplimiento), que puede tener los valores 'no', 'yes' o 'unknown'.


-  **Estadística Descriptiva**: Es el proceso de resumir y organizar datos para comprender sus características principales. Incluye medidas de tendencia central (media, mediana, moda) y medidas de dispersión (varianza, desviación estándar).

- **Estadística Inferencial**: Se utiliza para hacer inferencias o predicciones sobre una población a partir de una muestra de datos. Incluye pruebas de hipótesis, intervalos de confianza y regresión.

-  **Análisis Exploratorio de Datos (EDA)**: El proceso de examinar y visualizar datos para identificar patrones, tendencias y posibles valores atípicos. Gráficos, diagramas de dispersión y resúmenes estadísticos son herramientas comunes en EDA.

- **Validación y Limpieza de Datos**: La validación implica asegurarse de que los datos sean precisos y confiables, mientras que la limpieza implica eliminar valores atípicos, datos duplicados o inconsistencias en los datos..

- **Ética en el Análisis de Datos**: Comprender las implicaciones éticas de la recopilación y el uso de datos es crucial. Esto incluye la privacidad de los datos y la responsabilidad en el manejo de la información. # ¿Qué es el EDA?

En esta primera lección nos centraremos en el proceso de EDA, pero nos podríamos preguntar que significa esto "del EDA". Imaginemos que estamos con un grupo de amigos y nos dicen de ir a comer a un restaurante del que han oído hablar. Sin embargo, no te dicen nada más, no nos dicen qué tipo de comida sirven, cómo es el restaurante, el precio medio de la comida etc. En esta situación nos podemos ver algo desconcertadas con un montón de preguntas a las que queremos dar una respuesta para tomar una decisión.

En esta situación, la primera pregunta que nos podríamos hacer es: ¿qué tipo de comida sirven? o incluso ver algunas opiniones previas de otros clientes. Es decir, haremos un trabajo de investigación para saber si nos interesa ir o no ir a ese restaurante. Esto no es más que lo que hacen los analistas de datos, solo que nosotras lo llamaremos **Análisis Exploratorio de Datos** o **EDA**. 


El **Análisis Exploratorio de Datos** (EDA por sus siglas en inglés *Exploratory Data Analysis*) es una etapa fundamental en el proceso de análisis de datos. Consiste en examinar y visualizar los datos de manera sistemática con el objetivo de descubrir patrones, identificar tendencias, detectar valores atípicos y extraer información relevante para comprender mejor el conjunto de datos.

El EDA se realiza generalmente al principio de un proyecto de análisis de datos, antes de aplicar técnicas de limpieza más avanzadas. A través del análisis exploratorio, se busca familiarizarse con los datos, identificar posibles problemas de calidad, determinar la distribución y variabilidad de las variables, y generar hipótesis que puedan guiar el análisis posterior.

Algunas de las técnicas comunes utilizadas en el Análisis Exploratorio de Datos incluyen:

1. Resumen estadístico: Se calculan medidas descriptivas como la media, la mediana, la desviación estándar, el rango y los percentiles para comprender la distribución y variabilidad de los datos.

2. Visualización de datos: Se utilizan gráficos y diagramas, como histogramas, gráficos de dispersión, diagramas de caja, diagramas de barras y gráficos de línea, para visualizar las características de los datos y explorar las relaciones entre las variables.

3. Análisis de correlación: Se examina la correlación entre las variables para identificar posibles relaciones y dependencias. Esto se puede realizar mediante el cálculo de coeficientes de correlación, como el coeficiente de correlación de Pearson.

4. Tratamiento de valores atípicos: Se buscan valores atípicos o extremos que puedan tener un impacto significativo en los resultados del análisis. Estos valores pueden ser errores de medición o indicativos de fenómenos inusuales o interesantes.

5. Imputación de datos faltantes: Si hay datos faltantes en el conjunto de datos, se pueden utilizar técnicas para estimar o imputar los valores faltantes de manera que no afecten significativamente los resultados del análisis.

6. Segmentación y agrupación: Se pueden aplicar técnicas de segmentación o agrupación para identificar subgrupos o patrones particulares dentro del conjunto de datos.

> En resumen, el Análisis Exploratorio de Datos se refiere al proceso de realizar una serie de investigaciones inciales sobre los datos que tenemos para poder descubrir patrones, detectar anomalías, probar hipótesis y comprobar suposiciones con la ayuda de estadísticas y representaciones gráficas. 

Y es que es **ESENCIAL ENTENDER LOS DATOS** y tratar de **OBTENER el mayor número de CONOCIMIENTO** a partir de ellos. 

Sin embargo, es importante tener en cuenta que el análisis exploratorio de datos no es un conjunto de técnicas, pasos o reglas. **El análisis exploratorio es una "filosofía" sobre como se deben analizar los datos** dependiendo en gran parte de las preguntas que queramos responder. 

## ¿Cómo realizar un análisis exploratorio de datos?

Como hemos dicho, el EDA no tiene una receta universal que podamos seguir paso a paso. Entonces nos puede surgir la siguiente pregunta:

> **¿Cómo realizar un análisis exploratorio de datos?**
>
>En realidad, esta es una de las preguntas que todo el mundo desea conocer. 

La respuesta es que depende del conjunto de datos con el que se trabaje. No hay un método o unos métodos comunes para llevar a cabo EDA. 

A lo largo de esta lección vamos a tratar de entender algunos métodos comunes y los pasos que se utilizan en el proceso de EDA. Estas "pasos" podrían ser: 

- **Entender las variables/columnas**: en esta parte entenderemos qué variables tenemos, qué tipos de datos, si tenemos valores nulos o duplicados, qué valores únicos tenemos en nuestro *dataset*, etc.

- **Limpiar el *dataset***: una vez identificados todos los valores únicos, los duplicados y/o nulos procederemos a dejar nuestro *dataset* lo más limpio posible, es decir, quitar duplicados, nulos, columnas irrelevantes etc. 

- **Analizar las relaciones entre las variables**: nos permitirá encontrar patrones, relaciones o anomalías en nuestros datos que nos ayudarán a sacar conclusiones.

## Entender las variables/columnas del *dataset*

Se refiere a la tarea de investigar y comprender las columnas presentes en el conjunto de datos que estamos analizando. Algunas de las tareas que podríamos incluir en esta fase son:

1. **Identificar las variables**: En esta etapa, se examina el conjunto de datos para identificar todas las variables o columnas presentes. Se puede revisar la descripción del conjunto de datos o explorar los encabezados de las columnas para comprender qué aspectos se están registrando o midiendo.

2. **Tipos de datos**: Es importante comprender qué tipos de datos contiene cada variable. Algunas variables pueden ser numéricas, otras categóricas, binarias, ordinales o de tiempo, como se mencionó anteriormente. Esto es relevante para seleccionar las técnicas adecuadas de análisis y visualización posteriormente.

3. **Valores nulos o faltantes**: Durante esta etapa, se busca la presencia de valores nulos o faltantes en las variables. Los valores faltantes pueden surgir debido a errores en la recopilación de datos, problemas técnicos o simplemente porque algunos elementos no tienen valores registrados para ciertas variables. La identificación de valores faltantes es crucial para decidir cómo manejarlos posteriormente, ya sea eliminando filas con valores faltantes o realizando una imputación de datos para llenar los vacíos.

4. **Valores duplicados**: Se busca la presencia de valores duplicados en las variables. Esto implica verificar si hay registros idénticos o muy similares en las columnas del conjunto de datos. Los valores duplicados pueden afectar los análisis y generar resultados incorrectos, por lo que es importante detectarlos y decidir cómo manejarlos, ya sea eliminándolos o investigando su causa.

5. **Valores únicos**: Se analizan los valores únicos presentes en cada variable. Esto puede revelar información importante sobre la variabilidad y la distribución de los datos. Contar los valores únicos y su frecuencia puede ayudar a identificar patrones o anomalías, y proporcionar una visión general de la diversidad de los datos.

En resumen, al "entender las variables/columnas" durante el Análisis Exploratorio de Datos, nos enfocamos en examinar y comprender las características y propiedades de las variables presentes en el conjunto de datos. Esto nos permite obtener información inicial sobre la estructura y la calidad de los datos antes de realizar análisis más detallados y tomar decisiones sobre cómo manejar los datos faltantes, duplicados o únicos.

Genial!! Pero solo lo hemos hecho para una columna, en realidad no interesaría hacerlo para todas las variables categóricas del DataFrame, ya que lo que en realidad queremos es explorar todas las columnas categóricas. Por lo tanto, lo que tendremos que hacer es: 

- Seleccionar las variables que son categóricas de nuestro DataFrame, para esto usaremos el método `.select_dtypes()` que es una función en Pandas que nos permite seleccionar columnas de un DataFrame basándose en su tipo de datos. Devuelve un DataFrame que contiene solo las columnas que cumplen con los tipos de datos especificados en los argumentos `include` y `exclude`. La sintaxis básica del método `select_dtypes()` es la siguiente:

    ```python
    DataFrame.select_dtypes(include=None, exclude=None)
    ```

    - `include` (opcional): Especifica los tipos de datos que se deben incluir en la selección. Puede ser una cadena de texto o una lista de tipos de datos, como `'int'`, `'float'`, `'object'`, etc. Si no se proporciona este argumento, se seleccionarán todas las columnas que coincidan con los tipos de datos especificados en `exclude`.

    - `exclude` (opcional): Especifica los tipos de datos que se deben excluir de la selección. Puede ser una cadena de texto o una lista de tipos de datos. Si no se proporciona este argumento, se seleccionarán todas las columnas que coincidan con los tipos de datos especificados en `include`.


    Algunos ejemplos de cómo se puede utilizar el método `select_dtypes()`:

    ```python
    # Seleccionar todas las columnas numéricas
    df.select_dtypes(include=['int', 'float'])

    # Seleccionar todas las columnas de tipo objeto (cadenas de texto)
    df.select_dtypes(include='object')

    # Seleccionar todas las columnas excepto las numéricas
    df.select_dtypes(exclude=['int', 'float'])

    # Seleccionar todas las columnas excepto las de tipo objeto (cadenas de texto)
    df.select_dtypes(exclude='object')
    ```

## Valores nulos

Los valores nulos, también conocidos como valores faltantes o *missing values*, son aquellos que no están presentes en una variable o columna específica de un conjunto de datos. Estos valores pueden surgir por diversas razones, como errores en la recopilación de datos, problemas técnicos, falta de respuesta o simplemente porque algunos elementos no tienen valores registrados para ciertas variables.

Es importante hacer una buena identificación/gestión de los valores nulos por varias razones:

1. **Impacto en el análisis**: Los valores nulos pueden afectar negativamente el análisis de datos, ya que la mayoría de las técnicas y algoritmos requieren que todas las observaciones tengan valores completos. Si no se manejan adecuadamente, los valores nulos pueden generar resultados incorrectos o sesgados. Imaginad que estamos analizando los datos de ventas de una tienda minorista. Si no gestionas adecuadamente los valores nulos en la columna de "cantidad de productos vendidos", podríamos obtener resultados incorrectos al calcular la cantidad total de ventas, calcular el promedio de ventas diarias o realizar cualquier otro tipo de análisis. Si ignoramos los valores nulos, el análisis podría subestimar las ventas o sesgar los resultados, lo que llevaría a decisiones comerciales erróneas.

2. **Precisión y confiabilidad**: La presencia de valores nulos puede comprometer la precisión y confiabilidad de los resultados y conclusiones obtenidos a partir de los datos. Si se ignoran los valores nulos, se corre el riesgo de obtener estimaciones inexactas o interpretaciones incorrectas. Supongamos que estamos analizando datos de encuestas de satisfacción del cliente. Si no manejas los valores nulos en las respuestas de los clientes, podrías obtener estimaciones inexactas de la satisfacción general, ya que no tendrías en cuenta las opiniones de aquellos clientes que no respondieron ciertas preguntas. Esto podría llevar a conclusiones incorrectas sobre los aspectos que influyen en la satisfacción del cliente y, en última instancia, afectar la toma de decisiones para mejorar la calidad del servicio o producto.

3. **Integridad de los datos**: La gestión adecuada de los valores nulos ayuda a mantener la integridad general de los datos. Al identificar y manejar los valores nulos, se asegura que los datos sean coherentes y completos, lo que facilita su uso en análisis posteriores. Imaginemos que tenemos un conjunto de datos que contiene información sobre pacientes en un estudio clínico. Si no gestionaramos los valores nulos en la columna de "edad del paciente", podría haber inconsistencias en el análisis, ya que los valores nulos podrían afectar la validez de los cálculos basados en la edad, como la edad promedio del grupo de pacientes o las tasas de éxito del tratamiento según la edad. Al identificar y manejar adecuadamente los valores nulos, se garantiza que los datos sean coherentes y completos, lo que mejora la calidad y la fiabilidad de los análisis y las conclusiones.



En Python hay 5 tipos principales de valores nulos:

1. **NaN**: NaN, que significa "Not a Number", es el valor nulo más común en Pandas. Representa un valor faltante o no válido en una serie o columna de datos. En Pandas, NaN se considera como un valor de tipo float.

2. **None**: None es un objeto de Python que se utiliza para representar la ausencia de valor en un contexto no numérico. A diferencia de NaN, que se utiliza principalmente en contextos numéricos, None puede ser utilizado en cualquier tipo de dato en Pandas, incluyendo objetos, cadenas de texto, entre otros. None se considera un valor de tipo NoneType.

3. **valores texto**: a veces nos podemos encontrar con valores de tipo texto pero que no son valores reales, si no que nos indican que no hay valor. Algunos ejemplos de esto son: 
   
    -  `"n/a"`, `"NaN"`, `"nan"`, `"null"`: son NaN que serán literalmente *strings*. Esto puede ocurrir cuando se han metido en nuestra base de datos literalmente con esa estructura. Aún con todo, podemos vernos en dos situaciones: 
4. **Los 99999 o 00000**: es similar al caso anterior, pero en este caso vienen codificados como *integers*. Esto es usual encontrarlo en datos que vienen de sensores o sistemas de medición continua. 

5. **NaT**: los veremos cuando tengamos valores nulos en columnas de tipo `datetime`.

Una indentificación/gestión adecuada de los valores nulos implica tomar decisiones sobre cómo manejarlos. Este apartado lo no desarrollaremos ampliamente ya que tendremos una lección dedicada a la gestión de nulos en Pandas. Sin embargo, algunas de las estrategias más comunes que veremos son:

- Eliminación de filas o columnas con valores nulos: Si los valores nulos son pocos y no afectan significativamente el análisis, se pueden eliminar las filas o columnas correspondientes. Sin embargo, esta estrategia puede llevar a una pérdida de información si los valores nulos contienen información relevante.

- Imputación de valores: Consiste en estimar o reemplazar los valores nulos con valores calculados a partir de otros datos o utilizando técnicas de imputación, como el promedio, la mediana o modelos de imputación más complejos.

- Creación de variables indicadoras: En algunos casos, se pueden crear variables indicadoras o banderas que indiquen si un valor es nulo o no. 


### Métodos para identificar nulos en Pandas

En Pandas, puedes identificar los valores nulos en un DataFrame utilizando diferentes métodos y funciones. Aquí os presentamos las formas comunes de hacerlo:

1. **isnull()**: devuelve un DataFrame o una Serie booleana que indica si cada valor en el DataFrame es nulo o no. Los valores nulos se representan como `True`, mientras que los valores no nulos se representan como `False`.Se puede aplicar este método a un DataFrame completo o a una columna específica. Por ejemplo:

    ```python
    # Verificar si hay valores nulos en todo el DataFrame
    df.isnull()

    # Verificar si hay valores nulos en una columna específica
    df['columna'].isnull()
    ```

2. **notnull()**: Similar a `isnull()`, devuelve un DataFrame o una Serie booleana que indica si cada valor en el DataFrame no es nulo. Los valores no nulos se representan como `True` y los valores nulos como `False`. Al igual que `isnull()`, puedes aplicar este método a un DataFrame completo o a una columna específica.

    ```python
    # Verificar si hay valores no nulos en todo el DataFrame
    df.notnull()

    # Verificar si hay valores no nulos en una columna específica
    df['columna'].notnull()
    ```

3. **info()**: proporciona un resumen conciso de un DataFrame, incluyendo el número de valores no nulos en cada columna. Esto puede ser útil para identificar rápidamente las columnas que tienen valores nulos. Por ejemplo:

    ```python
    # Mostrar información resumida del DataFrame
    df.info()
    ```

4. **sum()**: Al llamar al método `sum()` en un DataFrame o una Serie booleana (como los que nos devuelve el método `.isnull()` o `.notnull()`), se puede obtener el recuento total de valores nulos en cada columna. El valor nulo se trata como `True` y se suma como 1, mientras que los valores no nulos se tratan como `False` y se suman como 0. Por ejemplo:

    ```python
    # Contar los valores nulos en cada columna del DataFrame
    df.isnull().sum()

    # Contar los valores nulos en una columna específica
    df['columna'].isnull().sum()
    ```

## Valores duplicados

Los valores duplicados se refieren a la presencia de registros/filas repetidos en un conjunto de datos. Esto significa que hay al menos una o más observaciones o filas que comparten exactamente los mismos valores en todas las columnas.

Existen varias razones por las cuales podemos tener valores duplicados en un conjunto de datos. Algunas de ellas son:

1. **Errores de entrada**: Durante la recopilación o inserción de datos, pueden ocurrir errores humanos o técnicos, lo que resulta en la duplicación involuntaria de registros.

2. **Fusión de conjuntos de datos**: Cuando se combinan diferentes fuentes de datos, puede haber duplicados entre los conjuntos originales, especialmente si no se lleva a cabo un proceso adecuado de eliminación de duplicados durante la fusión.

3. **Actualizaciones o modificaciones**: Si se realizan actualizaciones o modificaciones en un conjunto de datos existente, es posible que los valores duplicados se creen debido a cambios en los registros.

La gestión adecuada de los valores duplicados es esencial en el proceso de análisis de datos por varias razones:

1. **Precisión y validez de los resultados**: Los valores duplicados pueden afectar la precisión de los análisis y generar resultados sesgados o inexactos. Al eliminar los duplicados, nos aseguramos de trabajar con datos únicos y obtener resultados más confiables.  Imaginemos que estamos realizando un estudio sobre preferencias de consumo de alimentos en una encuesta. Si permitimos que los participantes envíen respuestas duplicadas, podrías obtener una imagen distorsionada de las preferencias, ya que algunos alimentos podrían parecer más populares de lo que realmente son. Al eliminar los duplicados, puedes obtener resultados más precisos y confiables para comprender las preferencias reales.

2. **Eficiencia en el procesamiento**: Los valores duplicados aumentan el tamaño y la complejidad del conjunto de datos, lo que puede ralentizar el procesamiento y consumo de recursos innecesarios. Al eliminar los duplicados, optimizamos el rendimiento y la eficiencia del análisis.  Supongamos que estamos analizando registros de ventas en una empresa de comercio electrónico con una gran cantidad de datos. Si hay valores duplicados en los registros de ventas, el procesamiento de esos datos llevará más tiempo y requerirá más recursos computacionales. Al eliminar los duplicados, optimizas el proceso de análisis, lo que conduce a un procesamiento más rápido y eficiente. 

3. **Consistencia y coherencia de los datos**: La presencia de valores duplicados puede generar inconsistencias en los datos y dificultar la interpretación correcta. Al gestionar los duplicados, garantizamos la coherencia de los datos y facilitamos su interpretación. Consideremos ahora una base de datos de empleados que incluye información personal, como nombres y direcciones. Si hay registros duplicados en la base de datos debido a errores de entrada, se puede generar inconsistencia en los datos. Por ejemplo, un empleado puede tener múltiples entradas con diferentes direcciones. Al gestionar los duplicados, se asegura la coherencia de los datos y se evita la confusión al interpretar la información correcta de cada empleado.

4. **Toma de decisiones correctas**: Los análisis se realizan con el propósito de tomar decisiones informadas. Los valores duplicados pueden distorsionar la información y llevar a decisiones erróneas. Al gestionar los duplicados, mejoramos la calidad de los datos y, por ende, la toma de decisiones. Supongamos que estamos realizando un análisis de satisfacción del cliente y deseamos identificar los problemas más frecuentes reportados. Si hay valores duplicados en las quejas registradas, es posible que los problemas se vean amplificados y se tomen decisiones basadas en información distorsionada. Al gestionar los duplicados, puedes obtener una visión más precisa y equilibrada de los problemas reales, lo que facilita la toma de decisiones correctas para abordarlos de manera efectiva.

¿Qué haremos con los duplicados cuando los tengamos? Aquí os mostramos algunas de las opciones, pero al igual que en el caso de los nulos, lo desarrollaremos apliamente más adelante en el bootcamp:

-  **Eliminar los duplicados**: Si los duplicados no son relevantes para el análisis y no aportan información adicional. 

-  **Agrupar y resumir**: En algunos casos, puede ser útil agrupar los datos por ciertas columnas y realizar resúmenes o agregaciones sobre los valores duplicados.

-  **Marcar y analizar**: En lugar de eliminar los duplicados, podemos agregar una columna adicional para marcar los registros duplicados y luego analizarlos por separado. 

- **Investigar la causa**: Si los duplicados son inesperados o indican posibles problemas en los datos de origen, es importante investigar y corregir la causa del problema. Esto puede implicar verificar los procesos de entrada de datos, la calidad de los datos de origen o solucionar cualquier problema técnico que pueda estar generando duplicados.


### Métodos para indentificar los duplicados en Pandas

En Pandas, solo podemos identificar los duplicados usando el método `.duplicated()` el cual va a identificar las filas duplicadas en un DataFrame y devolverá una Serie booleana que indica True para las filas duplicadas y False para las filas únicas.

Su sintaxis básica es:

```python
df.duplicated(subset=['columna1', 'columna2'], keep='first')
```

El método `duplicated()` tiene los siguientes parámetros principales:

- **subset** (opcional): Este parámetro permite especificar las columnas sobre las cuales se comprobarán los duplicados. Puedes pasar una lista de nombres de columnas para considerar solo esas columnas al buscar duplicados. Si no se especifica, se considerarán todas las columnas del DataFrame.

- **keep** (opcional): Este parámetro controla qué ocurrencias de los duplicados se marcarán como `True`. Los valores posibles son:

  - `'first'` (valor predeterminado): Marca como `True` todas las filas duplicadas, excepto la primera aparición de cada conjunto de duplicados.

  - `'last'`: Marca como `True` todas las filas duplicadas, excepto la última aparición de cada conjunto de duplicados.
  
  - `False`: Marca como `True` todas las filas duplicadas, incluyendo todas las apariciones de duplicados.

````python
# como hemos visto lo anterior nos devuelve una serie de booleanos lo cual es poco entendible, 
# por lo que añadiremos el método '.sum()' para que nos cuente cuantas filas tenemos duplicadas
# en este caso el código nos esta diciendo que tenemos 1994 filas duplicadas
df.duplicated().sum()
````

````python
# ¿Y si quisieramos saber si tenemos duplicados para una columna en concreto? 
# en ese caso tendremos que usar el parámetro subset()' 
# imaginemos que queremos saber si hay duplicados en la columna de 'date' 
# en este caso tenemos un total de 41309 fechas duplicadas en nuestro conjunto de datos (recordemos que teníamos 43170 datos)

df.duplicated(subset = "date").sum()
````
