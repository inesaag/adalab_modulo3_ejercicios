# Introducción 

Comenzamos una nueva lección y en esta ocasión, exploraremos los métodos de unión en Pandas, una herramienta fundamental para combinar y analizar conjuntos de datos relacionados. En nuestro caso, trabajaremos con dos DataFrames que contienen información personal de los clientes a lo largo de tres años, junto con un tercer DataFrame que detalla una campaña de marketing para impulsar las compras.

En el mundo del análisis de datos, a menudo nos encontramos con el desafío de trabajar con múltiples fuentes de información. Puede que tengamos datos históricos en diferentes períodos de tiempo, información adicional sobre los clientes o incluso campañas de marketing diseñadas para influir en su comportamiento de compra. Aquí es donde entran en juego los métodos de unión en Pandas.

En esta lección, aprenderemos cómo utilizar las funciones de unión, como *merge*, *concat* y *join*, para combinar nuestros DataFrames y obtener una visión completa de la información. Veremos cómo podemos utilizar columnas comunes como claves de unión para vincular los datos (igual que hacíamos en SQL con los *joins*) de manera significativa y realizar análisis más profundos.

A través de ejemplos prácticos y ejercicios interactivos, descubriremos cómo unir los DataFrames correspondientes a los diferentes años y combinarlos con la información de la campaña de marketing. Aprenderemos a identificar y resolver posibles desafíos que puedan surgir durante el proceso de unión, como datos faltantes, duplicados o diferentes estructuras de datos.

El uso de las funciones de combinación de datos en Pandas, como *merge*, *concat* y *join*, es muy útil para los analistas de datos por las siguientes razones:

1. Combinar datos de múltiples fuentes: En el análisis de datos, a menudo trabajamos con datos provenientes de diferentes fuentes, como bases de datos, archivos CSV, hojas de cálculo, etc. Las funciones de combinación de datos nos permiten unir o concatenar estos conjuntos de datos en una única estructura, lo que facilita su manipulación y análisis.

2. Ampliar conjuntos de datos: En ocasiones, necesitamos agregar información adicional a un conjunto de datos existente. Con merge, concat y join, podemos combinar diferentes conjuntos de datos utilizando columnas comunes como clave de unión, lo que nos permite agregar más información a nuestros datos.

3. Análisis de datos relacionales: Cuando trabajamos con datos relacionales, es decir, datos que están vinculados por una relación lógica, las funciones de combinación de datos son fundamentales. Nos permiten unir tablas o conjuntos de datos basándonos en claves relacionales, como identificadores únicos, y realizar análisis combinados, como agregaciones o filtrados, en función de esas relaciones.

4. Limpiar y transformar datos: Las funciones de combinación también son útiles para limpiar y transformar datos. Por ejemplo, podemos combinar conjuntos de datos para eliminar duplicados, filtrar registros basados en condiciones específicas, reordenar columnas o realizar cálculos derivados.

5. Integración de datos estructurados y no estructurados: En algunos casos, es posible que tengamos datos estructurados (por ejemplo, tablas) y datos no estructurados (por ejemplo, texto o imágenes) que necesitemos integrar para un análisis completo. Las funciones de combinación de datos nos permiten unir estos diferentes tipos de datos y realizar análisis más profundos y completos.

# Métodos de unión de datos

Como ya hemos comentado en Pandas tenemos tres métodos para unir datos, veamos una pequeña descripción antes de ponernos manos a la obra: 


- **Concat**: El método `concat` se utiliza para combinar dos o más DataFrames vertical u horizontalmente. Permite unir los DataFrames a lo largo de un eje (por defecto, el eje 0, que es el eje de las filas) para crear un nuevo DataFrame. No hay una columna de referencia específica para unir los DataFrames, simplemente se concatenan uno tras otro.

- **Merge**: El método `merge` se utiliza para combinar dos DataFrames basándose en una o más columnas comunes. Es similar a la operación de *join* en SQL. Se especifica una o más columnas clave para unir los DataFrames y se pueden seleccionar diferentes tipos de combinaciones, como *inner join*, *left join*, *right join* o *outer join*. El resultado es un nuevo DataFrame que combina las filas de los DataFrames de acuerdo con las columnas clave.

- **Join**: El método `join` también se utiliza para combinar dos DataFrames, pero se basa en las etiquetas de índice en lugar de columnas comunes. Puede unir DataFrames en función de los índices y realizar una combinación similar a la operación de *join* en SQL. Al igual que `merge`, se pueden seleccionar diferentes tipos de combinaciones.

Diferencias:

- `Concat`: Se utiliza para combinar DataFrames simplemente apilándolos uno encima del otro o uno al lado del otro, sin tener en cuenta las columnas específicas o las claves de unión.

- `Merge`: Se utiliza para combinar DataFrames en función de una o más columnas comunes. Es similar a la operación de unión en SQL.

- `Join`: Se utiliza para combinar DataFrames en función de las etiquetas de índice en lugar de columnas comunes. Se puede considerar como una variante de `merge` que se basa en los índices en lugar de las columnas.

## `concat()`

El método `concat()` se utiliza para concatenar (unir) objetos de pandas a lo largo de un eje específico(filas o columnas). Puede ser utilizado para concatenar DataFrames o Series de pandas. Su sintaxis es:

```python
pd.concat(objs, axis=0, join='outer', ignore_index=False)
```

Parámetros principales:
- `objs`: Es un iterable de objetos de pandas (por ejemplo, DataFrames, Series) que se desean concatenar. Se puede pasar como una lista o un diccionario de objetos.

- `axis`: Especifica el eje a lo largo del cual se va a realizar la concatenación. Sus valores pueden ser: 

    - `axis=0` (por defecto) le dice a pandas que apile el segundo DataFrame DEBAJO del primero. Es decir, estamos concatenando por **filas**.
        > Cuando usamos este eje, tenemos que asegurarnos de que tenemos las mismas columnas y el mismo formato de columna asociado en ambos conjuntos de datos. En caso de que tengamos alguna columna que no este en los dos dataframes, nos pondrá NaN en los valores para el *DataFrame* que no tenía esa columna.

    - `axis=1` apilará las columnas del segundo DataFrame a la DERECHA del primer DataFrame. Es decir, estamos concatenando por **columnas** .
        > Cuando usamos este eje, queremos asegurarnos de que lo que estamos haciendo tiene sentido (es decir, los datos están relacionados de alguna manera).

- `join` (opcional): Especifica cómo se deben manejar las columnas que no se encuentran en todos los objetos. Puede ser 'outer' (unión de todas las columnas), 'inner' (intersección de las columnas comunes) u otras opciones.

- `ignore_index` (opcional): Indica si se deben restablecer los índices originales y generar nuevos índices en el resultado concatenado. Por defecto es `False`.

Para poner en práctica este tipo de unión usaremos los tres DataFrames que hemos cargado desde el Excel. El objetivo será tener toda la información personal de todos los clientes en un único DataFrame.

## `merge()` 

Probablemente la forma más común de unir *DataFrames* basado en algunas condiciones comunes. Se utiliza para combinar (fusionar) dos o más DataFrames en función de una o más columnas comunes. Puede realizar operaciones de unión similar a las que se encuentran en las bases de datos relacionales como SQL. Su sintaxis es:

```python
df_left.merge(df_right, how='inner', on=None, left_on=None, right_on=None)
```
Donde: 

- `df_left` y `df_right`: Son los DataFrames que se van a combinar.

- `how` (opcional): Especifica el tipo de unión a realizar. Los valores comunes son: 
    - `left`: usa sólo las claves del *DataFrame* izquierdo, similar a una unión externa izquierda de SQL.

    - `right`: usa sólo las claves del *DataFrame* derecho, similar a una unión externa derecha de SQL.

    - `outer`: utiliza la unión de las claves de ambos *DataFrame*, es decir, me va a devolver todo. 
    
    - `inner`: utiliza la intersección de las claves de ambos *DataFrame*, similar a una unión interna de SQL.

    - `cross`: crea el producto cartesiano de ambos marcos, conserva el orden de las claves de la izquierda.


- `on`: Especifica la(s) columna(s) común(es) en la que se basará la fusión. Si las columnas tienen el mismo nombre en ambos DataFrames, se puede especificar como una cadena. Si las columnas tienen nombres diferentes, se puede especificar como una lista de nombres de columna correspondientes.

- `left_on` y `right_on` (opcional): Permiten especificar columnas diferentes en `left` y `right` en las que se basará la fusión.

- `suffixes` (opcional): Especifica los sufijos a agregar a las columnas que se solapan en `left` y `right`. Por defecto, se agrega `_x` a las columnas de `left` y `_y` a las columnas de `right`.


**¿Por qué hacemos esta diferenciación entre `on`, `left_on` y `right_on`?**

- Podemos usar `on` si los nombres de las columnas por las que queremos unir nuestros *datasets* se llaman igual. 

- Por el contrario, si las columnas por las que vamos a mergear nuestros *DataFrames* tienen distintos nombres usaremos `left_on` y `right_on`.
Para entenderlo mejor por defecto, el `merge` es como un *inner join* de SQL. 

En este caso uniremos el DataFrame creado en el primer ejemplo del concat que vimos previamente y el DataFrame de la campaña de marketing, el que cargamos desde el *csv*

### Inner merge 

Cuando se realiza un *inner merge*, también conocido como intersección, se combinan los datos de dos *DataFrames* en función de una o más columnas comunes, pero solo se conservan las filas donde las columnas coinciden en ambos DataFrames. Como tenemos una columna en común pero que se llaman diferente en los *DataFrames* usaremos el parámetro `right_on` y `left_on`.

````python
mergeado_inner = df_concat0.merge(df_mar, left_on='ID', right_on = "id_")
mergeado_inner.head()
````

````python
print(f"las filas y columnas que tenemos en el df concatenado son {df_concat0.shape}")
print("-------------------------------------------------------------------")

print(f"las filas y columnas que tenemos en el df de marketing son {df_mar.shape}")
print("-------------------------------------------------------------------")

print(f"las filas y columnas que tenemos en el mergeado_inner son {mergeado_inner.shape}")
print("-------------------------------------------------------------------")

````

Fijaos que ahora tenemos nuestros *datasets* unidos, se nos han juntado las columnas de los *DataFrames* en aquellos valores que coinciden los valores de `id_`, que es por donde hemos especificado que mergee. 

> Tenemos el mismo número filas en el *DataFrame* mergeado que en el de la campaña de marketing.  Esto es así porque hay 'id'  que están en `df_concat0` y no en `df_mar`. Como no encuentra el "match" los elimina. Es decir no se ha llamado a todos los clientes. 
>
>El *DataFrame* `mergeado_inner` resultante contiene todas las columnas de `df_concat` así como todas las columnas de `df_mar`.
>
>También observamos que la columna del 'id_' y 'ID', es decir las columnas por las que hemos mergeado están incluidas en *DataFrame* final, es decir, tenemos columnas redundantes. 

```python
# 📌en caso de que las columnas que vamos a unir se hubieran llamado igual tendríamos que hacer usado la siguiente sintaxis, usando el parámetro 'on', en vez de 'right_on' y 'left_on': 

mergeado_inner2 = df_2002.merge(df_species, on = "NOMBRE_COLUMNA_COMÚN")
mergeado_inner2.head()
```

### Left merge

Cuando se realiza un *left merge*, también conocido como unión izquierda, se combinan los datos de dos *DataFrames* en función de una o más columnas comunes, y se conservan todas las filas del *DataFrame* de la izquierda (el primer *DataFrame* especificado en la fusión), incluyendo las filas que no tienen coincidencias en el *DataFrame* de la derecha.

````python
mergeado_izq = df_concat0.merge( right=df_mar, how='left',
                        left_on='ID', right_on='id_')

mergeado_izq.head()
````

````python
print(f"las filas y columnas que tenemos en el df_concat son {df_concat0.shape}")
print("-------------------------------------------------------------------")


print(f"las filas y columnas que tenemos en el mergeado_izquierda son {mergeado_izq.shape}")
print("-------------------------------------------------------------------")
````

El *DataFrame* resultante de una *left join* se parece mucho al resultado del *inner join* en cuanto a las columnas que contiene. Sin embargo, a diferencia del *inner_join* (`mergeado_inner`), *left_join* (`mergeado_izq`) contiene el mismo número de filas que el *DataFrame* original de `df_concat0`. 

Cuando examinamos `mergeado_izq`, encontramos que hay filas en las que falta la información que debería proceder de `df_mar`. Es decir, se han generado 170 nulos en algunas columnas

> Tenemos 170 NaN , esto significa que hay 170 clientes que teníamos en el `df_concat0` que NO aparecían en `df_mar`, es decir, los clientes a los que no se contacto en la campaña de marketing. 

## `join()`

Se utiliza para combinar los datos de dos DataFrames en función de los índices de las filas. Proporciona una forma conveniente de realizar uniones basadas en índices en lugar de columnas.

La sintaxis básica del método `join()` es la siguiente:

```python
resultado = df_left.join(df_right, how='tipo_de_join',  lsuffix='', rsuffix='')
```

Donde:

- `df_left` y `df_right`: Son los *DataFrames* que se desean combinar.

- `how='tipo_de_join'` (opcional): Especifica el tipo de join que se va a realizar. Los tipos de join comunes son `'left'`, `'right'`, `'inner'` y `'outer'`, y se utilizan de manera similar a los tipos de join en el método `merge()`.

- `lsuffix` (opcional): Por defecto nada, en caso de que coincidan columnas en los DataFrames que queremos unir, será el sufijo que aparecerá en esas columnas en el DataFrame izquierdo.

- `rsuffix` (opcional): por defecto nada, en caso de que coincidan columnas en los DataFrames que queremos unir, será el sufijo que aparecerá en esas columnas en el DataFrame derecho.

El resultado del método `join()` es un nuevo DataFrame que combina los datos de `dataframe1` y `dataframe2` en función de los índices de las filas. Las filas se unen en función de los índices coincidentes, y las filas que no tienen coincidencias se excluyen del resultado final.

Es importante destacar que el método `join()` se basa en los índices de las filas, por lo que las columnas con los mismos nombres en ambos *DataFrames* se manejan automáticamente. ## Filtrado de datos

El filtrado de datos es una etapa fundamental en el proceso de limpieza de un *DataFrame* para los analistas de datos. Aquí os explicamos algunas razones por las cuales el filtrado de datos es importante en este contexto:

1. **Eliminación de datos irrelevantes**: Los *DataFrames* suelen contener una gran cantidad de datos, y muchos de ellos pueden no ser relevantes para el análisis que se va a realizar. Mediante el filtrado, se pueden eliminar las filas o columnas que no aportan información útil, lo que ayuda a reducir el ruido y a enfocarse en los datos que son realmente importantes.

2. **Tratamiento de valores faltantes**: Los valores faltantes, también conocidos como valores nulos o NaN, son comunes en los conjuntos de datos reales. El filtrado permite identificar y manejar adecuadamente estos valores faltantes. Puedes decidir eliminar las filas o columnas que contienen valores nulos si su ausencia no afecta significativamente al análisis, o puedes aplicar técnicas como la imputación para rellenar los valores faltantes con estimaciones adecuadas.

4. **Filtrado por condiciones específicas**: A menudo, los analistas de datos necesitan trabajar con subconjuntos de datos que cumplen ciertas condiciones o criterios. El filtrado permite seleccionar las filas que satisfacen una o varias condiciones específicas, lo que facilita el análisis focalizado en un grupo específico de datos de interés. Esto puede ayudar a obtener información más precisa y relevante para el análisis.

En Pandas, existen varios métodos que podremos utilizar para filtrar datos en un *DataFrame*. Aquí os presentamos algunos de los métodos más comunes:

1. **Operadores de comparación**: Se pueden utilizar operadores de comparación, como `>`, `<`, `>=`, `<=`, `==`, `!=`, para crear condiciones de filtrado basadas en los valores de una columna o varias columnas.

2. **Método `isin()`**: Se utiliza para filtrar filas en base a una lista de valores permitidos. Se utiliza para seleccionar las filas que contienen valores específicos en una columna.

3. **Método `between()`**: Se utiliza para filtrar filas en base a un rango de valores. Permite seleccionar las filas que tienen valores dentro de un rango especificado en una columna.

4. **Método `str.contains()`**: Se utiliza para filtrar datos basados en la presencia de una subcadena en una serie de objetos de tipo string. Este método devuelve un booleano que indica si cada elemento de la serie contiene o no la subcadena especificada.


### Operadores de comparación 

La forma de filtrado en pandas utilizando operadores de comparación es una técnica muy común y útil para seleccionar filas basadas en condiciones específicas. Aquí tenéis una explicación de cómo funciona:

1. **Crear una condición**: Primero, necesitaremos crear una condición utilizando uno o más operadores de comparación, como `>`, `<`, `>=`, `<=`, `==`, `!=`. Estos operadores se utilizan para comparar los valores de una columna o varias columnas con un valor o una expresión. 

2. **Aplicar la condición de filtrado**: Una vez que tenemos la condición, podemos aplicarla directamente dentro de corchetes `[]` al DataFrame. Esto devolverá solo las filas que cumplan la condición.

3. **Combinar condiciones**: También es posible combinar varias condiciones utilizando los operadores lógicos `&` (and) y `|` (or) para crear condiciones más complejas (como lo hacíamos en NumPy).

### Método `.isin()`

La función `isin()` en pandas se utiliza para filtrar datos en base a una lista de valores permitidos. Permite seleccionar filas en un *DataFrame* cuyos valores en una columna específica coinciden con cualquiera de los valores proporcionados en la lista.

Aquí tenéis una explicación de cómo usar `isin()` en el filtrado de pandas:

1. **Crear una lista de valores permitidos**: Primero, necesitaremos crear una lista que contenga los valores que deseamos permitir en el filtrado. Pueden ser valores únicos o una lista de varios valores. 

2. **Aplicar la función `isin()`**: Una vez que tenemos la lista de valores permitidos, podemos aplicar la función `isin()` a la columna deseada en el DataFrame utilizando corchetes `[]`.

Su sintaxis básica es: 
```python
dataframe[columna].isin(valores)
```
Donde:
- `dataframe`: Especifica el nombre del DataFrame sobre el que queremos aplicar el método. 

- `columna`: Especifica la columna del DataFrame en la que se aplicará el método.

- `valores`: Puede ser una lista, una Serie o un array-like que contiene los valores que se desean comprobar.

````python
df_final.head()
````

````python
df_final["education"].unique()
````

````python
# imaginemos que nos interesan solo los clientes que tienen un nivel de educación de 'basic 6y', 'basic 9y'. 
# podríamos usar lo métodos aprendidos previamente, pero la realidad es que cuantas más condiciones pongamos más probabilidad hay en equivocarnos. 
# para eso usaremos el método 'isin()' 

# primero definimos una lista con los valores que queremos filtrar
filtro_educacion = ['basic 6y', 'basic 9y']

# aplicamos el método 'isin()' pasando la lista
df_educacion = df_final[df_final["education"].isin(filtro_educacion)]

# comprobamos que ahora solo tenemos dos valores únicos para la columna ' education'
df_educacion["education"].unique()
````

### Método `.between()`

El método `between` en Pandas se utiliza para filtrar filas de un DataFrame o una Serie basándose en un rango numérico o de fechas. Comprueba si los valores de la columna seleccionada están dentro del rango especificado y devuelve una máscara booleana que indica qué filas cumplen con la condición.

La sintaxis del método `between` es la siguiente:
```python
dataframe[columna].between(inicio, fin, inclusive=True)
```
Donde:

- `dataframe`: Especifica el nombre del DataFrame sobre el que queremos aplicar el método. 

- `columna`: Especifica la columna del DataFrame en la que se aplicará el método.

- `inicio` y `fin`: Indican los límites del rango.

- `inclusive`: Es un parámetro opcional que determina si los límites del rango son inclusivos o exclusivos. Sus valores pueden ser:
    
    - both: incluye el inicio y el final. 

    - left: incluye el incio pero no el fin. 

    - right: incluye el fin pero no el inicio.

    - neither: no incluye el inicio ni el final.

### Método `.str.contains()`

El método `str.contains()` se utiliza en Python para verificar si un patrón de texto específico está presente en una serie de objetos de tipo cadena.

La sintaxis general del método `str.contains()` es la siguiente:
```python
dataframe[columna].str.contains(pat, case=True, na=nan, regex=True)
```

Donde:

- `dataframe`: Especifica el nombre del DataFrame sobre el que queremos aplicar el método. 

- `columna`: Especifica la columna del DataFrame en la que se aplicará el método.

- `pat`: Es el patrón de texto que deseas buscar en la columna.

- `case` (opcional):  Indica si se debe realizar una búsqueda sensible a mayúsculas y minúsculas. El valor predeterminado es `True`, lo que significa que se realiza una búsqueda sensible a mayúsculas y minúsculas.

- `na`(opcional): Lo tendremos que usar si nuestra columna tiene nulos. Nos ayuda a gestionar los nulos (si lo hay) de la columna. Toma dos valores True, si hay nulos los ignora y los incluye en el resultado; False, si hay nulos los ignora y no los incluye en el resultado.

- `regex` (opcional): Indica si el patrón dado debe tratarse como una expresión regular. El valor predeterminado es `True`, lo que significa que el patrón se interpreta como una expresión regular.

````python
# imaginemos que queremos trabajar con la columna de 'education' y quisieramos sacar todos aquellos clientes que en su educación aparece un dígito
# lo primero que debemos hacer es sacar el patrón de regex
patron_regex = "\d+"

df_numeros = df_final[df_final["education"].str.contains(patron_regex, regex = True, na = False)]
df_numeros.head()
````
