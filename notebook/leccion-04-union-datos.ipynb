# M칠todos de uni칩n de datos

Como ya hemos comentado en Pandas tenemos tres m칠todos para unir datos, veamos una peque침a descripci칩n antes de ponernos manos a la obra: 


- **Concat**: El m칠todo `concat` se utiliza para combinar dos o m치s DataFrames vertical u horizontalmente. Permite unir los DataFrames a lo largo de un eje (por defecto, el eje 0, que es el eje de las filas) para crear un nuevo DataFrame. No hay una columna de referencia espec칤fica para unir los DataFrames, simplemente se concatenan uno tras otro.

- **Merge**: El m칠todo `merge` se utiliza para combinar dos DataFrames bas치ndose en una o m치s columnas comunes. Es similar a la operaci칩n de *join* en SQL. Se especifica una o m치s columnas clave para unir los DataFrames y se pueden seleccionar diferentes tipos de combinaciones, como *inner join*, *left join*, *right join* o *outer join*. El resultado es un nuevo DataFrame que combina las filas de los DataFrames de acuerdo con las columnas clave.

- **Join**: El m칠todo `join` tambi칠n se utiliza para combinar dos DataFrames, pero se basa en las etiquetas de 칤ndice en lugar de columnas comunes. Puede unir DataFrames en funci칩n de los 칤ndices y realizar una combinaci칩n similar a la operaci칩n de *join* en SQL. Al igual que `merge`, se pueden seleccionar diferentes tipos de combinaciones.

Diferencias:

- `Concat`: Se utiliza para combinar DataFrames simplemente apil치ndolos uno encima del otro o uno al lado del otro, sin tener en cuenta las columnas espec칤ficas o las claves de uni칩n.

- `Merge`: Se utiliza para combinar DataFrames en funci칩n de una o m치s columnas comunes. Es similar a la operaci칩n de uni칩n en SQL.

- `Join`: Se utiliza para combinar DataFrames en funci칩n de las etiquetas de 칤ndice en lugar de columnas comunes. Se puede considerar como una variante de `merge` que se basa en los 칤ndices en lugar de las columnas.

## `concat()`

El m칠todo `concat()` se utiliza para concatenar (unir) objetos de pandas a lo largo de un eje espec칤fico(filas o columnas). Puede ser utilizado para concatenar DataFrames o Series de pandas. Su sintaxis es:

```python
pd.concat(objs, axis=0, join='outer', ignore_index=False)
```

Par치metros principales:
- `objs`: Es un iterable de objetos de pandas (por ejemplo, DataFrames, Series) que se desean concatenar. Se puede pasar como una lista o un diccionario de objetos.

- `axis`: Especifica el eje a lo largo del cual se va a realizar la concatenaci칩n. Sus valores pueden ser: 

    - `axis=0` (por defecto) le dice a pandas que apile el segundo DataFrame DEBAJO del primero. Es decir, estamos concatenando por **filas**.
        > Cuando usamos este eje, tenemos que asegurarnos de que tenemos las mismas columnas y el mismo formato de columna asociado en ambos conjuntos de datos. En caso de que tengamos alguna columna que no este en los dos dataframes, nos pondr치 NaN en los valores para el *DataFrame* que no ten칤a esa columna.

    - `axis=1` apilar치 las columnas del segundo DataFrame a la DERECHA del primer DataFrame. Es decir, estamos concatenando por **columnas** .
        > Cuando usamos este eje, queremos asegurarnos de que lo que estamos haciendo tiene sentido (es decir, los datos est치n relacionados de alguna manera).

- `join` (opcional): Especifica c칩mo se deben manejar las columnas que no se encuentran en todos los objetos. Puede ser 'outer' (uni칩n de todas las columnas), 'inner' (intersecci칩n de las columnas comunes) u otras opciones.

- `ignore_index` (opcional): Indica si se deben restablecer los 칤ndices originales y generar nuevos 칤ndices en el resultado concatenado. Por defecto es `False`.

Para poner en pr치ctica este tipo de uni칩n usaremos los tres DataFrames que hemos cargado desde el Excel. El objetivo ser치 tener toda la informaci칩n personal de todos los clientes en un 칰nico DataFrame.

## `merge()` 

Probablemente la forma m치s com칰n de unir *DataFrames* basado en algunas condiciones comunes. Se utiliza para combinar (fusionar) dos o m치s DataFrames en funci칩n de una o m치s columnas comunes. Puede realizar operaciones de uni칩n similar a las que se encuentran en las bases de datos relacionales como SQL. Su sintaxis es:

```python
df_left.merge(df_right, how='inner', on=None, left_on=None, right_on=None)
```
Donde: 

- `df_left` y `df_right`: Son los DataFrames que se van a combinar.

- `how` (opcional): Especifica el tipo de uni칩n a realizar. Los valores comunes son: 
    - `left`: usa s칩lo las claves del *DataFrame* izquierdo, similar a una uni칩n externa izquierda de SQL.

    - `right`: usa s칩lo las claves del *DataFrame* derecho, similar a una uni칩n externa derecha de SQL.

    - `outer`: utiliza la uni칩n de las claves de ambos *DataFrame*, es decir, me va a devolver todo. 
    
    - `inner`: utiliza la intersecci칩n de las claves de ambos *DataFrame*, similar a una uni칩n interna de SQL.

    - `cross`: crea el producto cartesiano de ambos marcos, conserva el orden de las claves de la izquierda.


- `on`: Especifica la(s) columna(s) com칰n(es) en la que se basar치 la fusi칩n. Si las columnas tienen el mismo nombre en ambos DataFrames, se puede especificar como una cadena. Si las columnas tienen nombres diferentes, se puede especificar como una lista de nombres de columna correspondientes.

- `left_on` y `right_on` (opcional): Permiten especificar columnas diferentes en `left` y `right` en las que se basar치 la fusi칩n.

- `suffixes` (opcional): Especifica los sufijos a agregar a las columnas que se solapan en `left` y `right`. Por defecto, se agrega `_x` a las columnas de `left` y `_y` a las columnas de `right`.


**쯇or qu칠 hacemos esta diferenciaci칩n entre `on`, `left_on` y `right_on`?**

- Podemos usar `on` si los nombres de las columnas por las que queremos unir nuestros *datasets* se llaman igual. 

- Por el contrario, si las columnas por las que vamos a mergear nuestros *DataFrames* tienen distintos nombres usaremos `left_on` y `right_on`.
Para entenderlo mejor por defecto, el `merge` es como un *inner join* de SQL.

Como tenemos una columna en com칰n pero que se llaman diferente en los *DataFrames* usaremos el par치metro `right_on` y `left_on`.

````python
mergeado_inner = df_concat0.merge(df_mar, left_on='ID', right_on = "id_")
mergeado_inner.head()
````

### Left merge

Cuando se realiza un *left merge*, tambi칠n conocido como uni칩n izquierda, se combinan los datos de dos *DataFrames* en funci칩n de una o m치s columnas comunes, y se conservan todas las filas del *DataFrame* de la izquierda (el primer *DataFrame* especificado en la fusi칩n), incluyendo las filas que no tienen coincidencias en el *DataFrame* de la derecha.

````python
mergeado_izq = df_concat0.merge( right=df_mar, how='left',
                        left_on='ID', right_on='id_')

mergeado_izq.head()
````

El *DataFrame* resultante de una *left join* se parece mucho al resultado del *inner join* en cuanto a las columnas que contiene. Sin embargo, a diferencia del *inner_join* (`mergeado_inner`), *left_join* (`mergeado_izq`) contiene el mismo n칰mero de filas que el *DataFrame* original de `df_concat0`.

## `join()`

Se utiliza para combinar los datos de dos DataFrames en funci칩n de los 칤ndices de las filas. Proporciona una forma conveniente de realizar uniones basadas en 칤ndices en lugar de columnas.

## Filtrado de datos

2. **Aplicar la condici칩n de filtrado**: Una vez que tenemos la condici칩n, podemos aplicarla directamente dentro de corchetes `[]` al DataFrame. Esto devolver치 solo las filas que cumplan la condici칩n.

### M칠todo `.isin()`

La funci칩n `isin()` en pandas se utiliza para filtrar datos en base a una lista de valores permitidos. Permite seleccionar filas en un *DataFrame* cuyos valores en una columna espec칤fica coinciden con cualquiera de los valores proporcionados en la lista.

Aqu칤 ten칠is una explicaci칩n de c칩mo usar `isin()` en el filtrado de pandas:

1. **Crear una lista de valores permitidos**: Primero, necesitaremos crear una lista que contenga los valores que deseamos permitir en el filtrado. Pueden ser valores 칰nicos o una lista de varios valores. 

2. **Aplicar la funci칩n `isin()`**: Una vez que tenemos la lista de valores permitidos, podemos aplicar la funci칩n `isin()` a la columna deseada en el DataFrame utilizando corchetes `[]`.

Su sintaxis b치sica es: 
```python
dataframe[columna].isin(valores)
```
Donde:
- `dataframe`: Especifica el nombre del DataFrame sobre el que queremos aplicar el m칠todo. 

- `columna`: Especifica la columna del DataFrame en la que se aplicar치 el m칠todo.

- `valores`: Puede ser una lista, una Serie o un array-like que contiene los valores que se desean comprobar.

````python
df_final.head()
````

````python
df_final["education"].unique()
````

````python
# imaginemos que nos interesan solo los clientes que tienen un nivel de educaci칩n de 'basic 6y', 'basic 9y'. 
# podr칤amos usar lo m칠todos aprendidos previamente, pero la realidad es que cuantas m치s condiciones pongamos m치s probabilidad hay en equivocarnos. 
# para eso usaremos el m칠todo 'isin()' 

# primero definimos una lista con los valores que queremos filtrar
filtro_educacion = ['basic 6y', 'basic 9y']

# aplicamos el m칠todo 'isin()' pasando la lista
df_educacion = df_final[df_final["education"].isin(filtro_educacion)]

# comprobamos que ahora solo tenemos dos valores 칰nicos para la columna ' education'
df_educacion["education"].unique()
````

### M칠todo `.between()`

El m칠todo `between` en Pandas se utiliza para filtrar filas de un DataFrame o una Serie bas치ndose en un rango num칠rico o de fechas. Comprueba si los valores de la columna seleccionada est치n dentro del rango especificado y devuelve una m치scara booleana que indica qu칠 filas cumplen con la condici칩n.

La sintaxis del m칠todo `between` es la siguiente:
```python
dataframe[columna].between(inicio, fin, inclusive=True)
```
Donde:

- `dataframe`: Especifica el nombre del DataFrame sobre el que queremos aplicar el m칠todo. 

- `columna`: Especifica la columna del DataFrame en la que se aplicar치 el m칠todo.

- `inicio` y `fin`: Indican los l칤mites del rango.

- `inclusive`: Es un par치metro opcional que determina si los l칤mites del rango son inclusivos o exclusivos. Sus valores pueden ser:
    
    - both: incluye el inicio y el final. 

    - left: incluye el incio pero no el fin. 

    - right: incluye el fin pero no el inicio.

    - neither: no incluye el inicio ni el final. 


````python
df_final.head(1)
````

### M칠todo `.str.contains()`

El m칠todo `str.contains()` se utiliza en Python para verificar si un patr칩n de texto espec칤fico est치 presente en una serie de objetos de tipo cadena.

La sintaxis general del m칠todo `str.contains()` es la siguiente:
```python
dataframe[columna].str.contains(pat, case=True, na=nan, regex=True)
```

Donde:

- `dataframe`: Especifica el nombre del DataFrame sobre el que queremos aplicar el m칠todo. 

- `columna`: Especifica la columna del DataFrame en la que se aplicar치 el m칠todo.

- `pat`: Es el patr칩n de texto que deseas buscar en la columna.

- `case` (opcional):  Indica si se debe realizar una b칰squeda sensible a may칰sculas y min칰sculas. El valor predeterminado es `True`, lo que significa que se realiza una b칰squeda sensible a may칰sculas y min칰sculas.

- `na`(opcional): Lo tendremos que usar si nuestra columna tiene nulos. Nos ayuda a gestionar los nulos (si lo hay) de la columna. Toma dos valores True, si hay nulos los ignora y los incluye en el resultado; False, si hay nulos los ignora y no los incluye en el resultado.

- `regex` (opcional): Indica si el patr칩n dado debe tratarse como una expresi칩n regular. El valor predeterminado es `True`, lo que significa que el patr칩n se interpreta como una expresi칩n regular.
